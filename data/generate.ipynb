```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from faker.providers.company.id_ID import Provider as IDCompanyProvider
from faker.providers.address.id_ID import Provider as IDAddressProvider
from faker.providers.phone_number.id_ID import Provider as IDPhoneNumberProvider
from faker import Faker
import re, random
from datetime import datetime, timedelta
```

```python
df_debtor = pd.read_csv("./data_nasabah.csv")
df_debtor
```

```python
list_kelamin = {0:'laki-laki', 1:'perempuan'}

list_occupation = {0:'Buruh', 1:'Pegawai Negeri', 2:'Pegawai Swasta', 3:'Profesional', 4:'Pengusaha'}

list_marital_status = {0:'Belum menikah',1:'Menikah', 2:'Cerai hidup', 3:'Cerai mati'}


data_nama_debtor = []
data_address_debtor = []
data_company_debtor = []
data_number_debtor = []
data_debtor_nik = []
data_zip_code = []
data_rt_debtor = []
data_rw_debtor = []
data_ttl = []
data_age = []
data_occupation = []
data_marital = []
data_gender = []

# Generate 5000 random names
fake = Faker()
coba = False
num_data = 5000
while coba == False:
    name = fake.name()
    data_nama_debtor.append(name)
    data_nama_debtor = list(set(data_nama_debtor))
    if len(data_nama_debtor) == num_data:
        break
    else:
        continue


# Function to generate unique random numbers
def generate_unique_random_numbers(n, start, end):
    unique_numbers = set()
    while len(unique_numbers) < n:
        num = random.randint(start, end)
        if num not in unique_numbers:
            unique_numbers.add(num)
            yield num

data_debtor_nik = generate_unique_random_numbers(num_data,1000000000000000,9999999999999999)
data_debtor_nik = list(data_debtor_nik)
data_debtor_nik = [str(number) for number in data_debtor_nik]


# Add the Indonesian address provider to the Faker generator
fake.add_provider(IDAddressProvider)

# Add the Indonesian company provider to the Faker generator
fake.add_provider(IDCompanyProvider)

# Add the Indonesian phone number provider to the Faker generator
fake.add_provider(IDPhoneNumberProvider)

for i in range(num_data):
    #Pembuatan variable debtor_Birth_place
    city_birth = fake.state()
    # Mendefinisikan tanggal awal dan akhir untuk rentang tanggal lahir yang diinginkan
    tanggal_awal = datetime(1960, 1, 1)
    tanggal_akhir = datetime(2000, 12, 31)
    # Menghasilkan tanggal lahir acak dalam rentang yang ditentukan
    tanggal_lahir_acak = tanggal_awal + timedelta(days=random.randint(0, (tanggal_akhir - tanggal_awal).days))
    # Mengonversi tanggal lahir acak menjadi string dengan format yang diinginkan
    tanggal_lahir_acak_string = tanggal_lahir_acak.strftime("%d-%m-%Y")
    data_debtor_ttl = city_birth + ", " + tanggal_lahir_acak_string
    data_ttl.append(data_debtor_ttl)

    # Pembauatan variable debtor_age
    tanggal_sekarang = datetime.now()
    # Hitung selisih tahun
    selisih_tahun = tanggal_sekarang.year - tanggal_lahir_acak.year
    data_age.append(selisih_tahun)

    #Pembuatan Variable debtor_martial_status
    random_status = random.choice(list(list_marital_status.values()))
    data_marital.append(random_status)

    #Pembuatan Variable debtor_occupation
    random_occupation = random.choice(list(list_occupation.values()))
    data_occupation.append(random_occupation)

    #Pembuatan Variable debtor_occupation
    random_gender = random.choice(list(list_kelamin.values()))
    data_gender.append(random_gender)


    #Pembuatan variable debtor_address
    debtor_addres = fake.address()
    data_address_debtor.append(debtor_addres)

    #pembuatan variable debtor_company
    debtor_company = fake.company()
    data_company_debtor.append(debtor_company)

    #pembuatan variable debtor_number
    debtor_number = fake.phone_number()
    # Pola regex untuk mencocokkan string yang dimulai dengan "+62"
    pattern = r'^\+62'
    try1 = True
    while try1 == True:
        if re.match(pattern, debtor_number):
            data_number_debtor.append(debtor_number)
            break
        else:
            debtor_number = fake.phone_number()
            continue

    #Pembuatan variable debtor_zip
    kata_alamat = debtor_addres.split()
    zip_kode = kata_alamat[-1]
    data_zip_code.append(zip_kode)

    #Pembuatan variable debtor_rt
    debtor_rt = random.randint(1, 20)
    data_rt_debtor.append(debtor_rt)

    #Pembuatan variable debtor_rw
    debtor_rw = random.randint(1, 20)
    data_rw_debtor.append(debtor_rw)


df_debtor['debtor_name'] = data_nama_debtor
df_debtor['debtor_address'] = data_address_debtor
df_debtor['debtor_company'] = data_company_debtor
df_debtor['debtor_number'] = data_number_debtor
df_debtor['debtor_nik'] = data_debtor_nik
df_debtor['debtor_zip'] = data_zip_code
df_debtor['debtor_rt'] = data_rt_debtor
df_debtor['debtor_rw'] = data_rw_debtor
df_debtor['debtor_birth_place'] = data_ttl
df_debtor['debtor_age'] = data_age
df_debtor['debtor_occupation'] = data_occupation
df_debtor['debtor_marital_status'] = data_marital
df_debtor['debtor_gender'] = data_gender

df_debtor.drop(["Unnamed: 0.2", "Unnamed: 0"], axis=1, inplace=True)
```
