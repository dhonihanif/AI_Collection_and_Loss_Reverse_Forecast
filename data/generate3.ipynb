```python
tenor = []
for i in range(len(df)):
    df.loc[i, "loan_amount"] = df.loc[i, "loan_amount"] * 2
    if df.loc[i, "net_income"] >= 25000000:
        if df.loc[i, "credit_risk"] == "Tertagih":
            tenor.append(np.random.randint(24, 36))
        else:
            tenor.append(np.random.randint(1, 24))
    elif df.loc[i, "net_income"] >= 15000000:
        if df.loc[i, "credit_risk"] == "Tertagih":
            tenor.append(np.random.randint(15, 26))
        else:
            tenor.append(np.random.randint(1, 15))
    elif df.loc[i, "net_income"] >= 5000000:
        if df.loc[i, "credit_risk"] == "Tertagih":
            tenor.append(np.random.randint(10, 20))
        else:
            tenor.append(np.random.randint(1, 12))
    else:
        if df.loc[i, "credit_risk"] == "Tertagih":
            tenor.append(np.random.randint(3, 12))
        else:
            tenor.append(np.random.randint(1, 3))

df["tenor"] = tenor
df.drop(["Unnamed: 0",
         "avg_amounts_previous_bills", "avg_previous_payment"], axis=1, inplace=True)

```
