```python
import numpy as np
import pandas as pd
from scipy.stats import shapiro
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
df = pd.read_csv(r"C:\Users\DHONI HANIF\OneDrive\Documents\AI Collection and Loss Reverse Forecast\data_preparation\EDA\Univariate_bivariate_multivariate\Univariate_Analysis\new_nasabah.csv")
df.head()
```

```python
df.info()
```

```python
for i in df.columns:
    if df[i].dtype == "object":
        df[i] = df[i].astype("category")
df.drop(["Unnamed: 0", "credit_risk"], axis=1, inplace=True)
df.info()
```

```python
df_ = pd.DataFrame()

col = {}
for i in df.columns:
    if df[i].dtype == "category":
        col[i] = LabelEncoder().fit(df[i])
        df_[i] = col[i].transform(df[i])
    else:
        df_[i] = df[i]
    

minmax = MinMaxScaler().fit(df_)
df_ = pd.DataFrame(minmax.transform(df_), columns=df_.columns)
    
df_
```

# Numeric Variable

```python
df_stat = {}
df_p_value = {}
for i in df.columns:
    if df[i].dtype == "int64" or df[i].dtype == "float64":
        stat, p_value = shapiro(df[i])
        alpha = 0.05
        # Hasil uji 
        df_stat[i] = stat
        # Interpretasi hasil
        if p_value > alpha:
            df_p_value[i] = f"Distribusi normal: {p_value}"
        else:
            df_p_value[i] = f"Distribui tidak normal: {p_value}"
```

```python
df_shapiro = pd.DataFrame()
df_shapiro["columns"] = [i for i in df_stat.keys()]
df_shapiro["stats"] = [df_stat[i] for i in df_stat]
df_shapiro["p_value"] = [df_p_value[i] for i in df_p_value]
df_shapiro
```
