```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_nasabah = pd.read_csv(r"C:\Users\DHONI HANIF\OneDrive\Documents\AI Collection and Loss Reverse Forecast\data_preparation\EDA\Univariate_bivariate_multivariate\Univariate_Analysis\new_nasabah.csv")
df_nasabah.info()
```

```python
for i in df_nasabah.columns:
    if df_nasabah[i].dtype == "object":
        df_nasabah[i] = df_nasabah[i].astype("category")
df_nasabah.drop("Unnamed: 0", axis=1, inplace=True)
df_nasabah.info()
```

Gender x education

```python
# Bivariate Analysis of Categorical Variables Gender vs Education
ax = sns.countplot(data=df_nasabah, x="debtor_education_level", hue="debtor_gender")
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment="right")
plt.legend(bbox_to_anchor=(1, 0.2))
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
# Scaling category variable

columns = {}
for i in df_nasabah.columns:
    if df_nasabah[i].dtype == "category" or df_nasabah[i].dtype == "object":
        columns[i] = LabelEncoder().fit(df_nasabah[i])
        df_nasabah[i] = columns[i].transform(df_nasabah[i])
 
fig, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(data=df_nasabah.corr(), annot=True, linewidths=.2, linecolor="white", cmap="winter", ax=ax)
```
