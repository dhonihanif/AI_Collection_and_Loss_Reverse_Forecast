```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from lazypredict.Supervised import LazyRegressor

bill_methods = LabelEncoder().fit(df_kolektor["avg_bill_methods"])
df_kolektor["avg_bill_methods"] = bill_methods.transform(df_kolektor["avg_bill_methods"])
df_kolektor.head(5)
```

```python
df_kolektor.drop(["time_to_collect", "total_cost"], axis=1)
```

```python
# X = df_kolektor.drop(["collect_sum", "total_cost", "bill_date"], axis=1)
scaler = MinMaxScaler().fit(df_kolektor)
X = pd.DataFrame(scaler.transform(df_kolektor), columns=df_kolektor.columns)
y1 = df_kolektor["time_to_collect"]
X
```

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test, y1_train, y1_test = train_test_split(X, y1, test_size=.2, random_state=42)
reg1 = LazyRegressor(verbose=0, ignore_warnings=True)
models, predictions = reg1.fit(X1_train, X1_test, y1_train, y1_test)
models
```

```python
y2 = df_kolektor["time_to_collect"]
X2_train, X2_test, y2_train, y2_test = train_test_split(X, y2, test_size=.2, random_state=42)
reg2 = LazyRegressor(verbose=0, ignore_warnings=True)
models, predictions = reg2.fit(X2_train, X2_test, y2_train, y2_test)
models
```
