```python
Q1 = np.quantile(df_nasabah["avg_amounts_previous_bills"], .25)
Q3 = np.quantile(df_nasabah["avg_amounts_previous_bills"], .75)

IQR = Q3 - Q1
min_IQR = Q1 - 1.5*IQR
max_IQR = Q3 + 1.5*IQR

low_outlier = [i for i in df_nasabah["avg_amounts_previous_bills"] if i < min_IQR]
high_outlier = [i for i in df_nasabah["avg_amounts_previous_bills"] if i > max_IQR]
for i in range(len(df_nasabah)):
    if df_nasabah.loc[i, "avg_amounts_previous_bills"] in low_outlier or df_nasabah.loc[i, "avg_amounts_previous_bills"] in high_outlier:
        df_nasabah.drop(i, axis=0, inplace=True)

df_nasabah.shape
        
```

```python
# Univariate Analysis of Numerical variables avg_amounts_previous_bills
fig, ax = plt.subplots(1, 2, figsize=(8, 5), constrained_layout=True)
sns.histplot(data=df_nasabah, x="avg_amounts_previous_bills", ax=ax[0])
sns.boxplot(data=df_nasabah, y="avg_amounts_previous_bills", ax=ax[1])
ax[0].set_title("avg_amounts_previous_bills")
ax[1].set_title("avg_amounts_previous_bills")
plt.show()
print(df_nasabah["avg_amounts_previous_bills"].describe())
```

```python
scaler = RobustScaler().fit_transform(df_nasabah["avg_amounts_previous_bills"].to_numpy().reshape(-1, 1))
fig, ax = plt.subplots(1, 2, figsize=(8, 5), constrained_layout=True)
sns.histplot(data=scaler, ax=ax[0])
sns.boxplot(data=scaler, ax=ax[1])
ax[0].set_title("avg_amounts_previous_bills")
ax[1].set_title("avg_amounts_previous_bills")
ax[0].set_xlim(scaler.min(), 3)
ax[1].set_ylim(-5, 5)
plt.show()
print(df_nasabah["avg_amounts_previous_bills"].describe())
```
